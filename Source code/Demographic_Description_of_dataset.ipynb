```python
import pandas as pd
import numpy as np
import statistics as st
from scipy.stats import sem
```

```python
# Load the data
df = pd.read_csv('https://raw.githubusercontent.com/shahriariit/opendataset/master/Parkinson%20(1).csv')
print(df.shape)
print(df.info())
```

```python
#Calculate Mean
df.mean()
```

```python
#Calculate Standard Error of Mean
df.sem()
```

```python
#Calculate Median
df.median()
```

```python
#Calculate mode
df.mode()
```

```python
#Calculate Standard deviation
df.std()
```

```python
#Calculate Skewness
df.skew()
```

```python
#Calculate Kurtosis
df.kurt()
```
